# NLP Preprocessing Pipeline 
## Objective
Build  robust text preprocessing system to clean noisy real-world text data.

```python
import re
from collections import Counter
```

## Task 1: Conceptual Understanding

### 1. Love vs love
"Love" and "love" are treated as different tokens unless we normalize case.

### 2. Without stopwords removal
Models become noisy and inefficient.

### 3. When stopword removal is harmful
- "not" changes sentiment
- question structure breaks

### 4. Stemming vs Lemmatization
Stemming cuts words, lemmatization returns meaningful root words.


```python
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)

    text = text.lower()
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    return tokens, " ".join(tokens)
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    print("\n---")
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    
```

```python
def analyze_tokens(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, round(avg_len, 2)
```

```python
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    total, unique, avg = analyze_tokens(tokens)

    print("\nSentence:", sentence)
    print("Total:", total, "Unique:", unique, "Avg Length:", avg)
```

```python
all_tokens = []

for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10:", counter.most_common(10))
print("Least 5:", counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print(result)
```

```python
edge_cases = ["", "hahaha", "123456", None]

for case in edge_cases:
    tokens, clean = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Clean:", clean)
```
